Q1. Write a Python program that generates a random integer between 1 and 100, and then prompts the user to guess the number. If the user's guess is higher than the actual number, the program should print "Too high, try again." If the guess is lower than the actual number, the program should print "Too low, try again." If the guess is correct, the program should print "Congratulations, you guessed the number!

The program should continue to prompt the user for guesses until they correctly guess the number. The program should also keep track of the number of guesses the user makes and print it out once they guess correctly. Finally, the program should ask the user if they want to play again, and if they do, generate a new random number and start the game over.

```python
import random #importing the random module from python library
print(f"--------------------------------------------------------------------------------------------------------------")
print(f"Welcome to Guess the number challenge!")
name = input("Enter your name: ") #asking the user to enter their name
print(f"--------------------------------------------------------------------------------------------------------------")
print(f"Welcome to the game",name)
print(f"--------------------------------------------------------------------------------------------------------------")
print(f"Rules and how to play:") #presenting the rules for the game
print(f"--------------------------------------------------------------------------------------------------------------")
print(f"> The computer will generate a random number between 1 to 100 and you have to find it")
print(f"> If you guess the number correctly, you win this game and you can choose to continue or exit.")
print(f"> If you guess a number that is lower than the number generated by the computer, then your guess will get stored until",
      "you","give a correct answer and then will be printed.")
print(f"> If you guess a number that is greater than the number generated by the computer, then your guess will be stored until"
     ,"you","give a correct answer and will be printed along with your other guesses.")
print(f"--------------------------------------------------------------------------------------------------------------")


begin = input("Would you like to begin the game?(Yes/No): ") #asking the user if they would like to begin their game
#using while loop, if the user enters 'yes' to begin their game in any of the forms mentioned below
while begin == 'yes' or begin == 'Yes' or begin == "YES":
    random_number = random.randint(1,100) #the computer will generate a random number from 1 to 100
    cont = 'y' #initialising a varaible cont='y'
    nolist = [] #creating an empty list to store the number of guesses made by the user until the right guess
#using while loop, if cont=='y'
#i.e., asking the user if they want to continue the game or not
    while cont =='y': 
        guess = int(input("Enter your guess of the number: ")) #asking the user to enter their guess
#using if condition, if the number guessed by the user is same as the number generated by the computer, then these lines get
#executed
        if guess == random_number:
            print("Congratulations, you guessed the number!!")
#printing the length of the list in which the guesses were stored, until the correct number was guessed
#i.e., the number of guessess made by the user until finding the correct number
            print(f"You guessed",{len(nolist)},"times.")
#printing all the guesses made by the user
            print(f"Your guesses were: ",nolist)
            break #breaking the if condition using break statement
        print(f"--------------------------------------------------------------------------------------------------------------")

        
        if guess < random_number: #checking if the number guessed by the user is lesser than the number generated by the computer
            nolist.append(guess) #appending this number onto the list 'nolist' if the above if condition is satisfied
            print("Too low, try again.") #printing this statement
        print(f"--------------------------------------------------------------------------------------------------------------")

        if guess > random_number: #checking if the number guessed by the user is greater than the number generated by the computer
            nolist.append(guess) #appending this number onto the list 'nolist' if the above if condition is satisfied
            print("Too high, try again.") #printing this statement
    cont = input("Would you like to continue?(y/n): ") #asking the user if they would like to continue to guess another number
    print(f"--------------------------------------------------------------------------------------------------------------")


    if cont == 'n': #if incase the user doesn't want to continue the game
        print(f"Okay. Thank You!") #printing this statement
        break #using the break statement and breaking the if condition
print(f"--------------------------------------------------------------------------------------------------------------")
```

Q2. Write a Python function called mysum that takes in a list of numbers as its argument and returns the sum of all the numbers in the list. The function should work in the same way as the built-in sum function, but you are not allowed to use the sum function in your implementation.

```python
def my_sum(lisht): #defining a function named my_sum having its argument to be a list of numbers
    l = len(lisht) #calculating the length of the list
#initialising s=0, since we need to calculate the sum of numbers inside the list mentioned as the argument in this function
    s = 0
    for i in range(0,l): #using for loop, for i in the range from 0 till the length of the list
        s = s+lisht[i] #calculating s by adding the index of the list with 0 and so on
    return s #returning the final value of 's', i.e., the sum of numbers in a list
```

```python
l1 = [1,2,3,4,5] #initialising a list of numbers
```

```python
my_sum(l1) #calling the function
```

Q3. Write a Python function called exercise_timing that asks the user how long it took them to complete a specific exercise routine. The function should then store the user's 
input in a list, and continue to prompt the user for more times until they enter "done". Once the user enters "done", the function should calculate and return the average time it took the user to complete the exercise routine.

```python
def excercise_timing(): #defining a function named 'excercise_timing'
    print(f"------------------------------------------------------------------------------------------------------")
    name = input("Enter your name: ") #asking the user to enter their name
    print(f"------------------------------------------------------------------------------------------------------")
    print(f"Hi",name,"! Welcome to the excercise time calculator!") #printing this statement
    print(f"------------------------------------------------------------------------------------------------------")
    list1 = [] #initialising an empty list
    while list1 == []: #using while loop, while the list is empty
        cont = 'yes'#initialising a variable cont='y'
        while cont == 'yes': #using while loop, if cont=='y'
#asking the user to enter the time taken to complete the excercise or enter done if they don't have any more timings to enter
            time = float(input("Enter the time taken to complete your excercise: "))
            list1.append(time) #appending the time asked above onto the empty list
            s = sum(list1) #calculating the sum of the list
            l = len(list1) #calculating the length of the list
#asking the user if they want to enter the time tsken to complete another excercise
            cont = input("Do u want to enter the time taken to complete another excercise?(yes/done): ")
            continue #using the continue statement and continuing the loop
            if cont == 'done': #checking if time=='done', using if condition
                print(f"------------------------------------------------------------------------------------------------------")
                avg_time = s/l #calculating the average time by dividing the sum by the length of the list
                print(f"Your average time to do excercise is:",avg_time) #printing this statement
                break #using the break statement, breaking the loop abruptly
        else: #an else for while
            average_time = s/l #calculating the average time by dividing the sum by the length of the list
    print(f"------------------------------------------------------------------------------------------------------")
    print(f"Your average time to complete an/the excercise(s) is: ",average_time) #printing this statement
    print(f"------------------------------------------------------------------------------------------------------")
        
excercise_timing() #calling the function to calculate the time to do the excercise
```

Q4. Write a Python function called hex_output that takes a hexadecimal number as its argument and returns the decimal equivalent of the number. The function should work for both lowercase and uppercase hexadecimal numbers, and should handle leading zeros in the input.

```python
def hex_output(hexa_num): #defining a function to calculate the decimal equivalent of a hexadecimal number
    hexa = hexa_num #initiating hexa equal to the haxadecimal number
    dechimal = int(hexa, 16) #converting the hexadecimal into its decimal equivalent
#printing the decimal equivalent for the hexadecimal number
    print(f'The decimal equivalent for the hexadecimal number {hexa_num} is  {dechimal}.')

print(f"--------------------------------------------------------------------------------------------------------------")
hexa_num = input("Enter you hexadecimal: ") #asking the user to enter the hexadecimal number
print(f"--------------------------------------------------------------------------------------------------------------")
hex_output(hexa_num) #calling the function to calculate the decimal equivalent for the hexadecimal number
print(f"--------------------------------------------------------------------------------------------------------------")
```

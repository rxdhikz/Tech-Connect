```python
import re

log_file_directory = "log_errors.txt"
timestamp_pattern = r'\[(.*?)\]'
error_pattern = r'ERROR: (.*?)\n'
total_errors = 0
distinct_errors = 0
da_file = open('log_errors.txt','r')
log_data = da_file.read()
timestamps = re.findall(timestamp_pattern,log_data)
errors = re.findall(error_pattern,log_data)
total_errors = len(errors)
distinct_errors = set(errors)

print(f"---------------------------------------------------------------------------------------------------------------------")
print(f"Log File Analyser")
print(f"---------------------------------------------------------------------------------------------------------------------")
print(f"The total number of errors are: ",total_errors)
print(f"The total number of distinct errors are: ",len(distinct_errors))
print(f"---------------------------------------------------------------------------------------------------------------------")
```
